```python
import cv2
from google.colab.patches import cv2_imshow
import requests
```

```python
url = "https://cdn.pixabay.com/photo/2016/11/29/10/07/tiger-1868911_1280.jpg"
r = requests.get(url)
with open ('test.jpg', 'wb') as f:
  f.write(r.content)
```

```python
img = cv2.imread('test.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
(thresh, binary) = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
cv2_imshow(img)
cv2_imshow(gray)
cv2_imshow(binary)
```
